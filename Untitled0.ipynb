```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/data/BitcoinHeistData.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
from tqdm.notebook import tqdm
labels = []
for i in tqdm(df['label']):
    if i=='white':
        labels.append(0)
    else:
        labels.append(1)
```

```python
x = df.drop(columns = 'label')
y = labels
```

```python
x.shape
```

```python
x.head()
```

```python
y
```

```python
x['address'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x['address'] = le.fit_transform(x['address'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

# Perceptron

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
p.fit(x_train_sc,y_train)
```

```python
train_pred_p = p.predict(x_train_sc)
test_pred_p = p.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy : ",accuracy_score(train_pred_p,y_train))
print("Testing Accuracy :",accuracy_score(test_pred_p,y_test))
```
